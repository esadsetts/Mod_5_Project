```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

# Data Exploration

Load in data

```python
df = pd.read_csv('data_cleaned.csv', index_col=0)
```

Plot average transfer value by year

```python
fig = plt.figure(facecolor = '#37474f')
ax = df.groupby('year').fee_cleaned.mean().plot(kind = 'bar', color = '#f5f5f5')
plt.xlabel('Year', color = '#f5f5f5')
plt.ylabel('Average Transfer Value', color = '#f5f5f5')
plt.xticks(color = '#f5f5f5')
plt.yticks(color = '#f5f5f5')
plt.title('Average Transfer Value by Year', color = '#f5f5f5')
ax.set_facecolor('#37474f')
```

Plot average transfer value by league

```python
fig = plt.figure(facecolor = '#37474f')
ax = df.groupby('league_name').fee_cleaned.mean().sort_values().plot(kind = 'bar',  color = '#f5f5f5')
plt.xlabel('League Name', color = '#f5f5f5')
plt.ylabel('Average Transfer Value', color = '#f5f5f5')
plt.xticks(color = '#f5f5f5')
plt.yticks(color = '#f5f5f5')
plt.title('Average Transfer Value by League', color = '#f5f5f5')
ax.set_facecolor('#37474f')
```

Plot average transfer value by age of player

```python
fig = plt.figure(facecolor = '#37474f')
ax = df.groupby('age').fee_cleaned.mean().plot(color = '#f5f5f5')
plt.xlabel('Age', color = '#f5f5f5')
plt.ylabel('Average Transfer Value', color = '#f5f5f5')
plt.xticks(color = '#f5f5f5')
plt.yticks(color = '#f5f5f5')
plt.title('Average Transfer Value by Age', color = '#f5f5f5')
ax.set_facecolor('#37474f')
```

Plot average transfer value by position

```python
fig = plt.figure(facecolor = '#37474f')
ax = df.groupby('position').fee_cleaned.mean().sort_values().plot(kind = 'bar', color = '#f5f5f5')
plt.xlabel('Position',  color = '#f5f5f5')
plt.ylabel('Average Transfer Value', color = '#f5f5f5')
plt.xticks(color = '#f5f5f5')
plt.yticks(color = '#f5f5f5')
plt.title('Average Transfer Value by Position', color = '#f5f5f5')
ax.set_facecolor('#37474f')
```

I plotted a histogram of all transfer fees

```python
fig = plt.figure(facecolor = '#37474f')
ax = sns.distplot(df.fee_cleaned, color = '#f5f5f5')
plt.title('Distribution of transfer values', color = '#f5f5f5')
plt.xlabel('Transfer Fee', color = '#f5f5f5')
plt.ylabel('Average Transfer Value', color = '#f5f5f5')
plt.xticks(color = '#f5f5f5')
plt.yticks(color = '#f5f5f5')
ax.set_facecolor('#37474f')
```
